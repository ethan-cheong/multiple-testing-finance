```python
from sklearn.linear_model import LinearRegression
import numpy as np

import statsmodels.api as sm
```

```python
def calculate_alpha(crsp_id):
    rets = df_all_long[df_all_long['crsp_fundno'] == crsp_id][['date', 'mret']].sort_values('date').reset_index()
    variables = pd.concat([rets, df_factors, df_tbill], axis=1).drop(['index', 'Date', '4 WEEKS COUPON EQUIVALENT'], axis=1)
    variables['mret'] = pd.to_numeric(variables['mret']) * 100
    variables['EXR'] = variables['mret'] - variables['RF']
    variables = variables.rename({"Mom   ":"UMD", "Mkt-RF" : "EXMKT"}, axis=1)[["EXMKT",  "HML", "SMB", "UMD", "EXR"]]
    X = variables[["EXMKT", "HML", "SMB", "UMD"]]
    X = sm.add_constant(X)
    y = variables[["EXR"]]
    model = sm.OLS(y, X)
    # Use HAC robust standard errors, since in the original paper, they did this
    # too. They checked the data using tests and found that only a few funds 
    # have homoskedasticity and temporal independence in their residuals which
    # is a required assumption for OLS.
    
    # Calculate maxlags term. Use ceil(0.75 * 36 ^ 1/3) rule of thumb
    
    results = model.fit(cov_type='HAC',cov_kwds={'maxlags':3})
    if crsp_id < 0:
        print(f"Results for model {crsp_id}:")
        print(results.summary())
    return results.params[0], results.params, results.resid, results.bse, results.rsquared, results.tvalues
    # return model.intercept_.item(), model.coef_, residualwe
```

# Bootstrap analysis 

Bootstrap analysis from https://onlinelibrary.wiley.com/doi/full/10.1111/j.1540-6261.2006.01015.x

```python
from tqdm import tqdm
```

```python
def draw_bootstrapped_alphas():
    # Takes index in the df_all df as an input.
    variables = pd.concat([df_factors, df_tbill], axis=1).drop(['Date', '4 WEEKS COUPON EQUIVALENT'], axis=1).rename({"Mom   ":"UMD", "Mkt-RF" : "EXMKT"}, axis=1)[["EXMKT",  "HML", "SMB", "UMD"]]
    new_alphas = []
    new_t_stats = []
    for i in range(len(df_all)-3): # exclude the summary funds
        alpha = alphas[i]
        coef_estimates = coefs[i]
        resids = residuals[i]
        t_stat = tvalues[i][0]
        resampled_resids = resids.sample(n = 36, replace=True).reset_index().drop("index", axis=1)[0]
        pseudo_excess_returns = coef_estimates['EXMKT'] * variables['EXMKT'] + coef_estimates['SMB'] * variables['SMB'] + coef_estimates['HML'] * variables['HML'] +coef_estimates['UMD'] * variables['UMD'] + resampled_resids 
        X = variables[["EXMKT", "HML", "SMB", "UMD"]]
        X = sm.add_constant(X)
        y = pseudo_excess_returns
        model = sm.OLS(y, X)
        results = model.fit()
        new_alphas.append(results.params[0])
        new_t_stats.append(results.tvalues[0])
    return new_alphas, new_t_stats
        
new_alphas = []
new_t_stats = []

for i in tqdm(range(1000)): # do 1000 bootstrap iterations
    new_alpha, new_t_stat = draw_bootstrapped_alphas()
    new_alphas.append(new_alpha)
    new_t_stats.append(new_t_stat)

new_alphas = np.array(new_alphas)
new_t_stats = np.array(new_t_stats)
new_t_stats_sorted = np.sort(new_t_stats, axis=0)
```

# Imposing model assumptions in the game-theoretic setting 

Four-factor Carhart model:
$$
r_{i,t} = \alpha_i + b_i \cdot r_{m,t} + s_i \cdot r_{smb, t} + h_i \cdot r_{hml, t} + m_i \cdot r_{mom, t} + \epsilon_{i,t}
$$

We want to test the null hypothesis that $$\alpha_i = 0$$ for each i. We assume that the model is properly specified, and use HAC-Robust Standard Errors. We can come up with estimates for alpha at each time step.
However, we write the test as a single linear restriction test, because we can then specify it as a normal distribution:

$$
H_0 : c' \beta = \gamma 
$$
where $$c = (1,0,0,0,0)^\top, \gamma = (0, 0,0,0,0)^\top$$ and the test statistic $$z = \frac{c'\hat\beta - \gamma}{se(c'\hat\beta - \gamma)} \sim ^a N(0,1)$$

Note that $$z = \frac{\hat\beta_0}{se\hat\beta_0}$$ which is just the linear regression t test.

```python
from scipy.stats import shapiro

def calculate_tvalue_at_times(crsp_id):
    # Use HAC-Robust standard errors.
    alphas = np.array([])
    shapiro_p_values = np.array([1, 1, 1])
    rets = df_all_long[df_all_long['crsp_fundno'] == crsp_id][['date', 'mret']].sort_values('date').reset_index()
    variables = pd.concat([rets, df_factors, df_tbill], axis=1).drop(['index', 'Date', '4 WEEKS COUPON EQUIVALENT'], axis=1)
    variables['mret'] = pd.to_numeric(variables['mret']) * 100
    variables['EXR'] = variables['mret'] - variables['RF']
    variables = variables.rename({"Mom   ":"UMD", "Mkt-RF" : "EXMKT"}, axis=1)[["EXMKT",  "HML", "SMB", "UMD", "EXR"]]
    for t in range(36): # ideally, we would use asymptotic theory here... 
        X = variables[["EXMKT", "HML", "SMB", "UMD"]][:t+1]
        X = sm.add_constant(X)
        y = variables[["EXR"]][:t+1]
        model = sm.OLS(y,X)
        results = model.fit(cov_type='HAC',cov_kwds={'maxlags':3})
        alphas = np.append(alphas, results.tvalues[0])
        resids = results.resid
        if t > 2:
            shapiro_p_values = np.append(shapiro_p_values, shapiro(resids)[1])        
    return alphas, shapiro_p_values

def get_alphas(crsp_id):
    return calculate_tvalue_at_times(crsp_id)[0]

def resid_normality(crsp_id):
    alpha, p_values = calculate_tvalue_at_times(crsp_id)
    return np.all(p_values > 0.05)
```

```python
def prep_df_using_ret_col(df_stock_all, ret_col):
    grouped_dates = df_stock_all.groupby("date").count().reset_index()
    new_dates = df_stock_all.query("date >= '1980-01-01'")
    max_count = new_dates.groupby("PERMNO").count()[ret_col].max()
    all_here = new_dates.groupby("PERMNO").count().reset_index().query(f"{ret_col} == {max_count}")['PERMNO']
    df_stock_filtered = df_stock_all[(df_stock_all['date'] >= '1980-01-01') & (df_stock_all["PERMNO"].isin(all_here))]
    df_stock_pivot = df_stock_filtered.pivot(index="PERMNO", columns ="date", values=[ret_col])
    df_stock_processed = df_stock_pivot.dropna(axis=0)
    dates = set()
    for i, date in df_stock_processed.columns:
        dates.add(date)
    
    dates = sorted(list(dates))
    
    index_to_date = {i+1 : date for i, date in enumerate(dates)}
    date_to_index = { ret_col + "_" + date : ret_col + "_" + str(i+1) for i, date in enumerate(dates)}
    df_stock_processed.reset_index(0).reset_index(drop=True)
    df_stock_processed.columns = df_stock_processed.columns.to_series().str.join('_')
    df_stock_processed = df_stock_processed.reset_index()
    df_stock_processed = df_stock_processed.dropna(axis=0)
    ret_cols = df_stock_processed.columns[df_stock_processed.columns.str.startswith(ret_col)]
    df_stock_processed = df_stock_processed.rename(mapper=date_to_index, axis=1)
    return df_stock_processed
```

Here, RET is return excluding dividends.

We are left with 449 stocks that have continuous return info from Jan 1980 to Dec 2023. This gives us 528 time periods, which should be enough time to reject if necessary.

```python
def add_cumprod(df_stock_processed, ret_col):
    # Make all returns non-negative
    df = df_stock_processed.copy()
    ret_cols = [col for col in df if col.startswith(ret_col)]
    df[ret_cols] =df[ret_cols] + 1
    # The cumulative product is an evalue!
    cumprod_cols = [f"cumprod_{i+1}" for i in range(len(ret_cols))]
    df[cumprod_cols] = df[ret_cols].cumprod(axis=1)
    return df
```

```python
def base_e_bh_1(col, alpha):
    # Returns the indices of the rejected evalues
    def R(t):
        n_evalues_greater_than_t = len(col[col >= t])
        return 1 if n_evalues_greater_than_t == 0 else n_evalues_greater_than_t
    t_alpha = float("inf")
    for t in range(1000000):
        if t * R(t) >= len(col) / alpha:
            t_alpha = t
            break
    return list(col[col >= t_alpha].index)
```

```python
def run_e_bh(df, colname, n_times, alpha):
    rejected = set()
    cols = [f"{colname}_{i+1}" for i in range(n_times)]
    rejected_against_time = []
    for col in cols:
        rejected.update(base_e_bh_1(df[col], alpha))
        rejected_against_time.append(len(rejected))
    return rejected_against_time
```

```python
run_e_bh(df, "cumprod", 528, 0.6425)
```

Interestingly, there seems to be a "hard threshold" that isn't a product of the number of time periods, or even the number of evalues! It probably depends on the relative magnitude / proportions of the evalues.

```python
def bisection_search(df, colname, n_times, a, b, tol, nmax):
    mapping = {}
    n = 1
    while n <= nmax:
        print(f"iteration number {n}")
        print(f"searching between {a} and {b}")
        c = (a + b) / 2
        if (b - a) / 2 < tol:
            return mapping
        n = n + 1
        attempt = run_e_bh(df, colname, n_times, c)
        mapping[c] = len(attempt)
        if len(attempt) == 0:
            a = c
        else:
            b = c
    return mapping
```

### Prepare to generate synthetic data: pull in factors

$$
r_{i,t} = \alpha_i + b_i \cdot r_{m,t} + s_i \cdot r_{smb, t} + h_i \cdot r_{hml, t} + m_i \cdot r_{mom, t} + \epsilon_{i,t}
$$

The generalised hyperbolic distribution is often used in economics, with particular application in the fields of modelling financial markets and risk management, due to its semi-heavy tails.

We use this distribution to model the alphas (there are sometimes firms with very consistently negative alphas, and also positive).

We choose this distribution because it empirically fits the dataset, and also is highly flexible allowing us to tweak the proportion of firms with 0 alpha

```python
def generate_synthetic_data(p, a, b, loc, scale, coefs, n_entries, fund_eps_stds, fund_eps_means):
    """
    Returns a tuple with both the data and the alphas, so we can examine which alphas are close to 0
    """
    print("reading and transforming data...")
    df_factors = pd.read_csv("../data/F-F_Research_Data_Factors.CSV")
    df_mom = pd.read_csv("../data/F-F_Momentum_Factor.CSV")
    # Get the last 1000 entries
    df_mom = df_mom.iloc[(len(df_mom) - n_entries):]
    df_factors = df_factors.iloc[(len(df_factors) - n_entries):]
    df_factors = pd.merge(df_factors, df_mom, on = "Date")
    np.random.seed(1)
    ## N_FUNDS = 2978
    df_factors["const"] = 1
    alphas = stats.genhyperbolic.rvs(p, a, b, loc, scale, size = 2978)
    # indices = np.random.choice(np.arange(alphas.size), replace=False,
    #                        size=int(alphas.size * prop_zero))
    # alphas[indices] = 0
    factors = df_factors.rename({"Mom   ":"UMD", "Mkt-RF" : "EXMKT"}, axis=1)[['const', 'EXMKT', 'HML', 'SMB', 'UMD']]
    factors["const"] = 1
    resids = []
    for i in tqdm(range(2978)):
        resids.append(np.random.normal(loc = fund_eps_means.iloc[i], scale = fund_eps_stds.iloc[i], size=n_entries))
    print("Joining resid array...")
    resids = np.array(resids)
    print("Doing regression matmul...")
    df_synth = (factors @ coefs).T + resids
    df_synth = df_synth.rename({i : f"ret_{i+1}" for i in range(n_entries)}, axis=1)
    return (df_synth, alphas)
```

```python
def convert_to_long(df_synth):
    df_long = df_synth.reset_index().rename({"index":"crsp_fundno"}, axis=1)
    df_long = pd.wide_to_long(df_long, stubnames = ["ret"], i = "crsp_fundno", j = "date", sep = "_")
    return df_long.reset_index()
```

```python
def calculate_alpha(crsp_id, df_long, df_factors):
    rets = df_long[df_long['crsp_fundno'] == crsp_id][['date', 'ret']].sort_values('date').reset_index()
    variables = pd.concat([rets, df_factors], axis=1).drop(['index', 'Date'], axis=1)
    variables['EXR'] = variables['ret'] - variables['RF']
    variables = variables.rename({"Mom   ":"UMD", "Mkt-RF" : "EXMKT"}, axis=1)[["EXMKT",  "HML", "SMB", "UMD", "EXR"]]
    X = variables[["EXMKT", "HML", "SMB", "UMD"]]
    X = sm.add_constant(X)
    y = variables[["EXR"]]
    model = sm.OLS(y, X)
    # Use HAC robust standard errors, since in the original paper, they did this
    # too. They checked the data using tests and found that only a few funds 
    # have homoskedasticity and temporal independence in their residuals which
    # is a required assumption for OLS.
    
    # Calculate maxlags term. Use ceil(0.75 * 36 ^ 1/3) rule of thumb
    
    results = model.fit(cov_type='HAC',cov_kwds={'maxlags':3})
    if crsp_id < 0:
        print(f"Results for model {crsp_id}:")
        print(results.summary())
    return results.params.iloc[0], results.params, results.resid, results.bse, results.rsquared, results.tvalues
    # return model.intercept_.item(), model.coef_, residual
```

```python
def experiment_1(df_synth, n_entries): 
    df_factors = pd.read_csv("../data/F-F_Research_Data_Factors.CSV")
    df_mom = pd.read_csv("../data/F-F_Momentum_Factor.CSV")
    # Get the last 1000 entries
    df_mom = df_mom.iloc[(len(df_mom) - n_entries):]
    df_factors = df_factors.iloc[(len(df_factors) - n_entries):]
    df_factors = pd.merge(df_factors, df_mom, on = "Date")
    np.random.seed(1)
    ## N_FUNDS = 2978
    df_long = convert_to_long(df_synth)

    alphas = []
    coefs = []
    residuals = []
    standard_errors = []
    rsquares = []
    tvalues = []
    print("Calculating regressions for each fund...")
    for i in tqdm(range(len(df_synth))):
        alpha, coef, residual, standard_error, rsquare, tvalue = calculate_alpha(i, df_long, df_factors)
        alphas.append(alpha)
        coefs.append(coef)
        residuals.append(residual)
        standard_errors.append(standard_error)
        rsquares.append(rsquare)
        tvalues.append(tvalue)
    return (alphas, coefs, residuals, standard_errors, rsquares, tvalues)
```
